```python
# Import pandas library
import pandas as pd
```

```python
# Load the dataset
df = pd.read_csv("Placement.csv")
```

```python
# Display first five rows
df.head()
```

```python
# Check missing values in each column
df.isnull().sum()
```

```python
# Fill missing salary values with 0 because students who are not placed do not have a salary.

df["salary"] = df["salary"].fillna(0)
```

```python
df.isnull().sum()
```

```python
df = df.drop_duplicates()
```

```python
print(df.shape)
```

```python
df["gender"] = df["gender"].str.strip().str.upper()

df["ssc_b"] = df["ssc_b"].str.strip().str.title()

df["hsc_b"] = df["hsc_b"].str.strip().str.title()

df["hsc_s"] = df["hsc_s"].str.strip().str.title()

df["degree_t"] = df["degree_t"].str.strip().str.title()

df["workex"] = df["workex"].str.strip().str.title()

df["specialisation"] = df["specialisation"].str.strip().str.title()

df["status"] = df["status"].str.strip().str.title()
```

```python
df.dtypes
```

```python
df["salary"] = df["salary"].astype(int)
```

```python
df.rename(columns={
    "ssc_p":"ssc_percentage",
    "hsc_p":"hsc_percentage",
    "degree_p":"degree_percentage",
    "etest_p":"employability_test_percentage",
    "mba_p":"mba_percentage"
}, inplace=True)
```

```python
df["average_percentage"] = (
    df["ssc_percentage"] +
    df["hsc_percentage"] +
    df["degree_percentage"] +
    df["mba_percentage"]
) / 4
```

```python
df.to_csv("cleaned_dataset.csv", index=False)
```
